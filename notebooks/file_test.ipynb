```python
from pathlib import Path
import pandas as pd
import os
```

```python
csv_folder = Path("../data/processed/football-data/PL")
```

```python
all_csvs = list(csv_folder.glob("matches_*.csv"))
```

```python
dfs = [pd.read_csv(f) for f in all_csvs]
```

```python
combined_df = pd.concat(dfs, ignore_index=True)
```

```python
print(f"Loaded {len(all_csvs)} season files")
print(f"Total rows: {len(combined_df)}")
print(combined_df.columns)
print(combined_df.head())
```
